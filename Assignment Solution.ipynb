# Analysis

1. Find the average number of calls made by a user in an org 				
				
Expected Output:				
Org ID,	no. of users, avg calls made by an user		
				


```python
query = ''' select org_id, count(user_id), avg(call_count) from 
(select org_id, user_id, count(*) as call_count
from call_logs group by org_id, user_id) t
group by org_id order by org_id
'''

pd.read_sql(query, connection)
```

2. Find out the number of users  who have atleast 3 connected calls in a month and who are given permission to see the contact number of the lead on a monthly basis.							
							
Expected Output: 						
Month, Active Orgs,	Users with atleast 3 connected calls and permission to display contact number					

```python
query = '''with t1 as (select month(str_to_date(call_date, '%d-%m-%Y')) as month, org_id, user_id as users, 
sum(call_connected)
from call_logs group by org_id, user_id, month having sum(call_connected)>=3
order by user_id, month), 
t2 as(select month, t1.org_id, users, permission_id from t1 left join user u on t1.users=u.user_id where permission_id=6) 

select month, t2.org_id, status, users from t2 join org o on t2.org_id = o.org_id''' 

pd.read_sql(query, connection)
```

```python
query = '''with t1 as (select month(str_to_date(call_date, '%d-%m-%Y')) as month, org_id, user_id as users, 
sum(call_connected)
from call_logs group by org_id, user_id, month having sum(call_connected)>=3
order by user_id, month), 
t2 as(select month, t1.org_id, users, permission_id from t1 left join user u on t1.users=u.user_id where permission_id=6) 

select month, status as `Active Orgs`, count(users) from t2 join org o on t2.org_id=o.org_id where status= 1 
group by 1, 2 order by month ''' 

pd.read_sql(query, connection)
```

3. Find the nunber of orgs that have enabled follow ups mandatory permission for their users. Also find and compare the number of calls made by orgs that have enabled the permission versus the orgs that have disabled the permission.												
												
Exopected Output:												
No of orgs enabled permission, Average no of calls made, No of orgs disabled permission,	Average no of calls made									

```python
query = '''with t as (select o.org_id, next_followup_mandatory, count(call_date) as call_cnt
from org o join call_logs c on o.org_id = c.org_id group by 1,2)

select count(org_id) as orgs_enabled_permission, 
avg(call_cnt) as avg_calls_made,
(select count(org_id) from t where next_followup_mandatory='False') as orgs_enabled_permission ,
(select avg(call_cnt) from t where next_followup_mandatory='False') as avg_calls_made
from t where next_followup_mandatory='True' '''

pd.read_sql(query, connection)
```

4. Find out the number of active orgs that have enabled or disabled each of the app settings listed in the Org Table				
				
Expected Output:				
* app_setting,	no of orgs enabled, no of orgs disabled		
* next_followup_mandatory				
* enable_automatic_dialling				
* view_leads_of_other_users				
* automatically_dial_after_sec 				
* display_start_calling_button				
* display_contact_number_on_lead 				

```python
query = '''with active_orgs as (select * from org where status=1)

select 'next_followup_mandatory' as app_setting, 
count(case when next_followup_mandatory = 'True' then 1 end) as orgs_enabled,
count(case when next_followup_mandatory = 'False' then 1 end) as orgs_disabled
from active_orgs
union 
select 'enable_automatic_dialling' as app_setting, 
count(case when enable_automatic_dialling = 'True' then 1 end) as orgs_enabled,
count(case when enable_automatic_dialling = 'False' then 1 end) as orgs_disabled
from active_orgs
union
select 'view_leads_of_other_users' as app_setting, 
count(case when view_leads_of_other_users = 'True' then 1 end) as orgs_enabled,
count(case when view_leads_of_other_users = 'False' then 1 end) as orgs_disabled
from active_orgs
union
select 'display_start_calling_button' as app_setting, 
count(case when display_start_calling_button = 'True' then 1 end) as orgs_enabled,
count(case when display_start_calling_button = 'False' then 1 end) as orgs_disabled
from active_orgs
union
select 'display_contact_number_on_lead' as app_setting, 
count(case when display_contact_number_on_lead = 'True' then 1 end) as orgs_enabled,
count(case when display_contact_number_on_lead = 'False' then 1 end) as orgs_disabled
from active_orgs
'''


pd.read_sql(query, connection)
```
